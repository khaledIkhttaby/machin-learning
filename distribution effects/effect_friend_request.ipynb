```python
import numpy as np
```

```python
def bot_is_already_found(bot_id=0,st=0,end=0):
    return False
def get_number_bots_for_every_range_in_distribution_normal(mean,variance,number_bots):
    res={}
    
    res[str(mean-variance)+","+str(mean)]=int(number_bots*0.34)
    res[str(mean)+","+str(mean+variance)]=int(number_bots*0.34)

    res[str(mean+variance)+","+str(mean+2*variance)]=int(number_bots*0.13)
    res[str(mean-2*variance)+","+str(mean-variance)]=int(number_bots*0.13)
    
    res[str(mean+2*variance)+","+str(mean+3*variance)]=int(number_bots*0.02)
    res[str(mean-3*variance)+","+str(mean-2*variance)]=int(number_bots*0.02)
  
    res[str(mean+3*variance)+","+str(mean+5*variance)]=int(number_bots*0.01)
    res[str(mean-5*variance)+","+str(mean-3*variance)]=int(number_bots*0.01)
    
    return res
def set_normal_distribution_for_bots(mean,variance,bots):
    result=get_number_bots_for_every_range_in_distribution_normal(mean,variance,len(bots))
    normal_dist=np.random.normal(mean, variance, size=(1, len(bots)))
    j=0
    set_result={}
    for k,v in result.items():
        try:
            
#             temp_normal_dist=normal_dist[normal_dist>int(k.split(",")[0])]
#             temp_normal_dist=normal_dist[normal_dist<int(k.split(",")[1])]
            for i in range(v):
                if bot_is_already_found():
                    j+=1
                    continue
                value=np.random.choice(np.arange(int(k.split(",")[0]),int(k.split(",")[1])))
                set_result[bots[j]]=value
                j+=1
        except Exception as e:
            print("----error----",e)
    return set_result
result=set_normal_distribution_for_bots(300,50,bots)
result
```
